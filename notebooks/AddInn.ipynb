```python
import re

import pandas as pd
import requests
```

```python
df = pd.read_excel("list_broker_insurance_mfo.xlsx", index_col=0)
df.head()
```

# Broker

```python
broker_df = pd.read_excel(
    "https://www.cbr.ru/vfs/finmarkets/files/supervision/list_brokers.xlsx", index_col=0, skiprows=3
)
broker_df.head()
```

```python
broker_df["licence"] = broker_df["№ лицензии"].replace({"-": ""}, regex=True).astype(int)
```

```python
df = df.join(broker_df.set_index("licence")["ИНН"], on="licence")
```

# mfo

```python
mfo_df = pd.read_excel(
    "list_MFO.xlsx",
    index_col=0,
    skiprows=4,
    # sheet_name="Исключенные",
    dtype={
        "Регистрационный номер записи": str,
        "Unnamed: 2": str,
        "Unnamed: 3": str,
        "Unnamed: 4": str,
        "Unnamed: 5": str,
    },
)
mfo_df["Регистрационный номер записи"] = mfo_df["Регистрационный номер записи"].fillna("")
mfo_df.head()
```

```python
mfo_df["licence"] = (
    mfo_df["Регистрационный номер записи"]
    + mfo_df["Unnamed: 2"]
    + mfo_df["Unnamed: 3"]
    + mfo_df["Unnamed: 4"]
    + mfo_df["Unnamed: 5"]
)
```

```python
mfo_df.tail()
```

```python
df = df.join(mfo_df.set_index("licence")["Идентификационный номер налогоплательщика"], on="licence", rsuffix="new_mfo")
```

# Insurance

```python
insurance_df = pd.read_excel("https://www.cbr.ru/vfs/finmarkets/files/supervision/list_ssd.xlsx", skiprows=2)
insurance_df.head()
```

```python
insurance_df = insurance_df[["Наименование субъекта страхового дела", "Лицензия", "ИНН"]].dropna()
insurance_df["licence"] = [int(re.findall(r"\d+", s)[0]) for s in insurance_df["Лицензия"].values]
```

```python
insurance_df.head()
```

```python
df = df.join(insurance_df.set_index("licence")["ИНН"], on="licence", rsuffix="insurance")
```

```python
df.to_excel("parsed_df.xlsx")
```
