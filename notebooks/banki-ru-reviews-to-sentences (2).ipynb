```python
import os

import numpy as np
import pandas as pd
import tqdm
```

```python
result = pd.read_csv("/home/samoed/Desktop/CourseWork3/data/banki.csv")
result.head()
```

```python
result.feeds = result.feeds.str.replace("([\.\?\!])([А-Я])", "\\1 \\2", regex=True)
result.head()
```

```python
result.columns.values[0] = "ID"
```

```python
import re

regex = re.compile(r"https?[:a-zA-Z0-9_.+-/#~?%]+")

step = 1000
for start in tqdm.tqdm(range(0, result.shape[0], step)):
    end = start + step
    result.feeds.iloc[start:end] = result.feeds.iloc[start:end].replace(regex, "URL")
result.head()
```

```python
preprocessed_path = "/home/bu/Science/Data/banki.ru/preprocessed/"
result.to_csv(preprocessed_path + "banki.ru", index=False)
```

```python
import nltk
import nltk.data

nltk.download("punkt")
tokenizer = nltk.data.load("tokenizers/punkt/russian.pickle")
```

```python
import tqdm

for_bert = None
file_num = 0

for r in tqdm.tqdm(result.index):
    r = result.loc[r]
    res = pd.DataFrame(
        {
            "ID": r[0],
            "sentence": tokenizer.tokenize(r.feeds),
            "weight": 0,
            "grades": r.grades,
        }
    )
    res.weight = 1.0 / res.shape[0]
    if for_bert is None:
        for_bert = res
    else:
        res.columns = for_bert.columns
        for_bert = pd.concat([for_bert, res], ignore_index=True)
    if for_bert.shape[0] > 100000:
        for_bert.to_csv(
            preprocessed_path + str(file_num) + "-sentences.csv", index=False
        )
        for_bert = None
        file_num = file_num + 1
```
