```python
import requests
import re
import pandas as pd
```

# Banks

```python
response = requests.get("https://www.cbr.ru/banking_sector/credit/FullCoList/")
text = response.text
```

```python
pattern = re.compile(r'<a\s+class="b-export_button"\s+title="Экспортировать в XLSX"\s+href="([^"]+)"')
```

```python
match = pattern.search(text)
match.group(1)
```

```python
url = "https://www.cbr.ru" + match.group(1).replace("amp;", "")
url
```

```python
content = requests.get(url).content
df = pd.read_excel(content)
df.head()
```

# MFO

```python
excel = pd.ExcelFile("https://www.cbr.ru/vfs/finmarkets/files/supervision/list_MFO.xlsx")
```

```python
excel.sheet_names
```

```python
types = {"Регистрационный номер записи": str, "Unnamed: 2": str, "Unnamed: 3": str, "Unnamed: 4": str, "Unnamed: 5": str}
working = excel.parse("Исключенные", skiprows=2, index_col=0, dtype=types)
working.head()
```

```python
working["Регистрационный номер записи"] = working["Регистрационный номер записи"].fillna('0')
working.info()
```

```python
working["licence"] = (
    working["Регистрационный номер записи"]
    + working["Unnamed: 2"]
    + working["Unnamed: 3"]
    + working["Unnamed: 4"]
    + working["Unnamed: 5"]
).astype(int)
working.tail()
```

```python
working.head()
```

```python
js = requests.get("https://www.sravni.ru/proxy-organizations/organizations/?limit=400&organizationType=mfo&skip=0").json()
```

```python
{item['status'] for item in js['items']}
```

```python
licences = [int(item['license']) for item in js["items"]]
```

```python
working[working["licence"].isin(licences)].head()
```
